```python
import matplotlib.pyplot as plt
import pandas as pd

# Define a custom MAD function(mean absolute deviation)
def mad(series):
    return (series - series.median()).abs().median()

# Group by 'Date' and calculate MAD
mad_df_canopy = df.groupby('Date')['Canopy Damper Position'].apply(mad)

# Sort by date
mad_df_canopy = mad_df_canopy.sort_index()

# Prepare for plotting
date_labels = mad_df_canopy.index.astype(str)
y_positions = range(len(mad_df_canopy))

# Plot
plt.figure(figsize=(15, 10))
plt.plot(mad_df_canopy.values, y_positions, label='MAD', color='purple')
plt.scatter(mad_df_canopy.values, y_positions, color='purple')

plt.yticks(ticks=y_positions, labels=date_labels)
plt.xlabel('MAD of Canopy Damper Position')
plt.ylabel('Date')
plt.title('Daily Median Absolute Deviation (MAD) of Canopy Damper Position')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your existing data
avg_s1 = df.groupby('Date')['Canopy Damper Position'].median()
avg_s1 = avg_s1.sort_index()

# Calculate the median
median_val = np.median(avg_s1.values)

# Prepare labels and positions for y-axis
date_labels = avg_s1.index.astype(str)
y_positions = range(len(avg_s1))  # Evenly spaced y-values

# Plot
plt.figure(figsize=(15, 10))
plt.plot(avg_s1.values, y_positions, label='Avg Canopy Damper Position')
plt.scatter(avg_s1.values, y_positions)

# Plot median as horizontal line across the x-axis range
plt.axvline(median_val, color='red', linestyle='--', label=f'Median = {median_val:.2f}')

plt.yticks(ticks=y_positions, labels=date_labels)
plt.xlabel('Avg Canopy Damper Position')
plt.ylabel('Date')
plt.title('Daily Avg Canopy Damper Position with Median')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
